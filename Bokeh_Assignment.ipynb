**Q1. How can you create a Bokeh plot using Python code?**

```python
# The Bokeh plot can be created using Python code as follows :
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
from bokeh.plotting import figure,output_file,show
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,8,6,2]
p = figure(title='Example of Bokeh Plot')
output_file('example.html')
p.xaxis.axis_label = 'X-axis'
p.yaxis.axis_label = 'Y-axis'
p.line(x,y)
show(p)
```

**Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.**

1. Glyphs are building blocks of Bokeh visualization.
2. A glyph is a graphical vectorized graphical shape or marker that is used to represent your data.
3. Different categories of glyphs are :

> marker - shapes like circles,diamonds,squares and triangles.

```python
from bokeh.models.markers import glyphs
x = [1,2,3,4,5,6]
y = [4,2,3,8,6,2]
p = figure(title='Example of Bokeh Plot')
output_file('example.html')
p.xaxis.axis_label = 'X-axis'
p.yaxis.axis_label = 'Y-axis'
p.circle(x,y)
show(p)
```

**Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?**

```python
x = [1,2,3,4,5,6]
y = [4,2,3,8,6,2]
p = figure(title='Example of Bokeh Plot')
output_file('example.html')
p.xaxis.axis_label = 'X-axis'
p.yaxis.axis_label = 'Y-axis'
p.line(x,y,line_color='#800080',legend_label='Data Point')
show(p)
# We can customize the appearance of Bokeh plot by including axis labels and legend labels and by filling the line colors in line graph and marker colors in marker plot
```

**Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?**


Here are the steps on how to embed a Bokeh plot into a web page or dashboard using Flask or Django:

Create a Bokeh plot.
Generate the HTML code that embeds the Bokeh plot.
Use Flask or Django to serve the HTML code.

from flask import Flask
from bokeh.embed import components

app = Flask(__name__)

@app.route("/")
def index():
    fig = bokeh.plotting.figure(title="My Plot")
    fig.scatter(x=[1, 2, 3], y=[4, 5, 6])
    script, div = components(fig)
    return render_template("index.html", script=script, div=div)

if __name__ == "__main__":
    app.run(debug=True)
